```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
data_filename = "amazon_data_18-11-19clean2.csv"
df = pd.read_csv(data_filename,delimiter=";")
df
colors = ("red","blue")
plt.bar(df['name'],df['seller_rating'],alpha=0.5)
plt.title('BAR PLOT')
plt.xlabel('Products')
plt.xticks(rotation=90)
plt.xticks(df['name'], ['IFB Microwave Oven',
       'Samsung Gaming Monitor',
       'OnePlus 7 Pro' ,
       'Apple iPad ',
       "Adidas Men's Running Shoes",
       'Gigabyte Mechanical Keyboard ',
       'Samsung 108 cm 4K TV',
       "M-WATCH Maxi 42 Men's Watch"])
plt.ylabel('Star Rating')
plt.ylim(4.4,5.2)
plt.show()
```
